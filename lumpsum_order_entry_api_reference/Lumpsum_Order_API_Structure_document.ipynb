# Lumpsum Purchase & Redemption Order Entry API

This notebook documents the **Lumpsum Purchase and Redemption Order Entry API** provided by **BSE Star MF Web Services**.

This API allows a mutual fund distributor or member company to place **lumpsum mutual fund transactions** through the BSE Star MF platform.

The API supports two main transaction types:

- **Purchase Order** → Buying mutual fund units
- **Redemption Order** → Selling mutual fund units

The request is sent to the BSE server using **SOAP XML web service**, and the server processes the order and returns a response indicating whether the transaction was successful.

This notebook contains screenshots of the official API documentation to explain the request structure, parameters, and response format.

# Purpose of the API

The **Lumpsum Order Entry API** is used by financial institutions, brokers, and distributors to place mutual fund orders through the BSE Star MF platform.

Using this API, the distributor backend system can send order details to BSE servers for processing.

The API allows the following actions:

- Purchase mutual fund units
- Redeem mutual fund units
- Submit transaction details
- Receive order confirmation

This API plays a critical role in enabling **automated mutual fund transaction processing**.

# API Workflow

The following steps describe how the Lumpsum Order Entry API works.

1. The **client or distributor backend system** collects transaction details.
2. The system prepares a **SOAP request message** containing order parameters.
3. The request is sent to the **BSE Star MF server**.
4. The server validates the request and processes the order.
5. The server generates a **transaction response**.
6. The response is sent back to the **client backend system**.

This process allows automated and secure processing of mutual fund transactions.

# API Request Structure

The Lumpsum Purchase and Redemption API requires several parameters to process a transaction.

These parameters include:

- Transaction details
- Client details
- Scheme information
- Authentication credentials
- Order amount or quantity
- Contact information
