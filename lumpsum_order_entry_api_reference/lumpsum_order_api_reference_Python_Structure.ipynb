# Lumpsum Purchase & Redemption Order Entry API

This notebook provides a **Python reference structure** for the BSE Star MF Lumpsum Order Entry API.

The code here is meant to help developers understand:

- API endpoint
- Request parameters
- Authentication requirements
- Expected response structure

```python
# ---------------------------------------------------
# API BASIC INFORMATION
# ---------------------------------------------------

API_NAME = "Lumpsum Purchase & Redemption Order Entry API"

API_METHOD = "orderEntryParam"

API_ENDPOINT = "https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure"

SOAP_ACTION = "http://bsestarmf.in/MFOrderEntry/orderEntryParam"

print("API Name:", API_NAME)
print("API Method:", API_METHOD)
print("API Endpoint:", API_ENDPOINT)
```

# Request Parameters

The following parameters must be included when sending an order entry request.

```python
# ---------------------------------------------------
# REQUEST PARAMETERS
# ---------------------------------------------------

REQUEST_PARAMETERS = {

"TransCode": "Transaction Code (NEW / MOD / CXL)",

"TransNo": "Unique transaction number generated by member",

"OrderId": "Order ID generated by BSE",

"UserID": "User ID provided by BSE",

"MemberId": "Member ID assigned by BSE",

"ClientCode": "Unique client identifier",

"SchemeCd": "Scheme code for mutual fund",

"BuySell": "Transaction type (P = Purchase, R = Redemption)",

"BuySellType": "Fresh or additional investment",

"DPTxn": "Demat transaction type",

"OrderVal": "Transaction amount",

"Qty": "Redemption quantity",

"AllRedeem": "Flag for full redemption",

"FolioNo": "Folio number for the investor",

"Remarks": "Optional remarks",

"KYCStatus": "Client KYC verification status",

"RefNo": "Reference number",

"SubBrCode": "Sub broker code",

"EUIN": "Employee Unique Identification Number",

"EUINVal": "EUIN declaration",

"IPAdd": "Client IP address",

"Password": "Encrypted password from authentication API",

"PassKey": "Random alphanumeric pass key",

"MobileNo": "Client mobile number",

"EmailID": "Client email address",

"MandateID": "Mandate ID for payment"
}

REQUEST_PARAMETERS
```

```python
# ---------------------------------------------------
# RESPONSE STRUCTURE
# ---------------------------------------------------

RESPONSE_STRUCTURE = {

"TransCode": "Transaction Code",

"TransNo": "Transaction Number",

"OrderNo": "Generated Order Number",

"UserID": "User ID",

"MemberId": "Member ID",

"ClientCode": "Client Code",

"Remarks": "Server message",

"SuccessFlag": "Indicates whether transaction was successful"

}

RESPONSE_STRUCTURE
```

# Summary

The Lumpsum Order Entry API enables distributors to automate mutual fund purchase and redemption transactions through the BSE Star MF platform.

The API requires proper authentication, transaction parameters, and scheme details.
