# SPREAD Order Entry API Structure

This notebook documents the structure of the **SPREAD Order Entry API** used in BSE StAR MF Web Services.

API Method:
**spreadOrderEntryParam**

**Purpose:**
This API allows members to place spread orders where redemption from one scheme and purchase into another scheme are processed together.

These transactions are processed as overnight spread orders.

The request is sent from the member system to the BSE StAR MF server using SOAP Web Services.

# Introduction to Spread Orders

A Spread Order allows an investor to switch investment from one mutual fund scheme to another.

This involves:

• Redemption from an existing scheme  
• Purchase into a new scheme  

The BSE StAR MF system processes these transactions sequentially during the overnight order cycle.
