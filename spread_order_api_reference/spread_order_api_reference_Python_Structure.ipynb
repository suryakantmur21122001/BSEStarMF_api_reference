# SPREAD Order Entry API – Python Reference

This notebook provides the Python reference structure for the SPREAD Order Entry API used in the BSE StAR MF Web Services.

**API Method**: **spreadOrderEntryParam**

**Purpose:**
This API allows members to place spread orders where redemption from one mutual fund scheme and purchase into another scheme are processed together.

The request is sent from the member system to the BSE StAR MF server using SOAP Web Services.

**Process Overview:**

- The member system authenticates using the Authentication API.
- The system generates an encrypted password and passkey.
- The spread order request is prepared with required parameters.
- The request is sent to the BSE StAR MF server.
- The server validates the request and registers the spread order.
- A response is returned containing the order ID and confirmation message.

**This notebook contains:**

• API basic information  
• Request parameter structure  
• Response structure  
• Sample SOAP request  
• Sample SOAP response

# API Basic Information:- 

```python
# -------------------------------------
# API BASIC INFORMATION
# -------------------------------------

API_NAME = "SPREAD Order Entry API"

API_METHOD = "spreadOrderEntryParam"

API_ENDPOINT = "https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure"

SOAP_ACTION = "http://bsestarmf.in/MFOrderEntry/spreadOrderEntryParam"

print("API NAME", API_NAME)
print("API METHOD", API_METHOD)
print("API ENDPOINT", API_ENDPOINT)
print("SOAP ACTION", SOAP_ACTION)
```

# Sample SOAP Request

```python
# -------------------------------------
# SAMPLE SOAP REQUEST
# -------------------------------------

SOAP_REQUEST_SAMPLE = """
<soap:Envelope xmlns:soap="http://www.w3.org/2003/05/soap-envelope"
xmlns:bses="http://bsestarmf.in/">

<soap:Header xmlns:wsa="http://www.w3.org/2005/08/addressing">
<wsa:Action>http://bsestarmf.in/MFOrderEntry/spreadOrderEntryParam</wsa:Action>

<wsa:To>
https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure
</wsa:To>

</soap:Header>

<soap:Body>

<bses:spreadOrderEntryParam>

<bses:TransactionCode>NEW</bses:TransactionCode>
<bses:UniqueRefNo>202308010000010</bses:UniqueRefNo>
<bses:SchemeCode>AXFCFDP-DP-L0</bses:SchemeCode>

</bses:spreadOrderEntryParam>

</soap:Body>

</soap:Envelope>
"""

print(SOAP_REQUEST_SAMPLE)
```
