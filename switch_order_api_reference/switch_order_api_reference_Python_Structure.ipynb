# SWITCH Order Entry API Python Reference Structure

This notebook provides a Python reference structure for the **SWITCH Order Entry API** used in the **BSE Star MF Web Services platform**.

The SWITCH Order Entry API allows members to switch an investment from one mutual fund scheme to another scheme.

In a switch transaction:

• Units are redeemed from the **source scheme**  
• The redeemed amount is invested into the **target scheme**

The API request is sent from the **member backend system** to the **BSE Star MF server** using **SOAP Web Services**.

API Method used: **switchOrderEntryParam**

# Summary

The SWITCH Order Entry API allows investors to transfer their investment from one mutual fund scheme to another scheme within the BSE Star MF platform.

The process involves two operations:

• Redemption of units from the source mutual fund scheme  
• Purchase of units in the target mutual fund scheme

The API request is sent from the member system to the BSE Star MF server using SOAP Web Services.

The request contains transaction details such as client information, scheme codes, transaction type, and authentication credentials.

After processing the request, the BSE server returns a SOAP response containing the order confirmation details including the unique order ID and transaction status.
