# SWITCH Order Entry API Structure

This notebook documents the structure of the **SWITCH Order Entry API** used in **BSE Star MF Web Services**.

**API Method:** `switchOrderEntryParam`

**Purpose:**  
This API is used to perform a **Switch transaction** where an investor switches their investment from one mutual fund scheme to another scheme.

In a SWITCH transaction:

• Units from one scheme are **redeemed (Switch Out)**  
• The amount is **invested into another scheme (Switch In)**

The SWITCH order is submitted from the **member system to the BSE Star MF server using SOAP Web Service**.

# Introduction to SWITCH Orders

A **Switch Order** allows an investor to move their investment from one mutual fund scheme to another scheme within the same fund house.

Instead of redeeming and purchasing separately, the system processes both operations together.

Example:

• Switching from **Equity Fund → Debt Fund**  
• Switching from **Regular Plan → Direct Plan**

Switch transactions are commonly used for **portfolio rebalancing or strategy changes**.

# SWITCH Request Method

The method used for SWITCH order request is:

`switchOrderEntryParam`

This method sends the switch transaction details from the **member system to BSE Star MF server**.

# Summary

The SWITCH Order Entry API allows members to move investments from one mutual fund scheme to another scheme through a single transaction.

This API performs two operations together:

• Redemption from the existing scheme  
• Purchase into the new scheme

The API request is sent using **SOAP Web Services**, and the response confirms the success or failure of the switch transaction.
