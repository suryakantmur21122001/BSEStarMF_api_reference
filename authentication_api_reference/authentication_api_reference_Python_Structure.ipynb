# BSE Star MF Authentication API Reference (Python Code Reference)

## API Overview

This notebook provides a reference structure for the BSE Star MF Authentication API.

API Method: **getPassword**

Purpose:
Authenticate the client system with the BSE Star MF server and generate an encrypted password (session ID) used for further API calls.

```python
# ----------------------------------
# API INFORMATION
# ----------------------------------

API_NAME = "Authentication API"
API_METHOD = "getPassword"

API_ENDPOINT = "https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure"

SOAP_ACTION = "http://bsestarmf.in/MFOrderEntry/getPassword"

print("API Name:", API_NAME)
print("API Method:", API_METHOD)
print("API Endpoint:", API_ENDPOINT)
```

```python
# ----------------------------------
# REQUEST PARAMETERS
# ----------------------------------

REQUEST_PARAMETERS = {
    "UserId": "Login ID provided by BSE",
    "Password": "Password for API login",
    "PassKey": "Random alphanumeric value generated by client"
}

REQUEST_PARAMETERS
```

```python
# ----------------------------------
# SAMPLE SOAP REQUEST
# ----------------------------------

SOAP_REQUEST = """
<soap:Envelope xmlns:soap="http://www.w3.org/2003/05/soap-envelope"
xmlns:bses="http://bsestarmf.in/">
<soap:Header xmlns:wsa="http://www.w3.org/2005/08/addressing">
<wsa:Action>http://bsestarmf.in/MFOrderEntry/getPassword</wsa:Action>
<wsa:To>
https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure
</wsa:To>
</soap:Header>

<soap:Body>
<bses:getPassword>
<bses:UserId>{{LoginID}}</bses:UserId>
<bses:Password>{{Password}}</bses:Password>
<bses:PassKey>{{Passkey}}</bses:PassKey>
</bses:getPassword>
</soap:Body>
</soap:Envelope>
"""

print(SOAP_REQUEST)
```

```python
# ----------------------------------
# RESPONSE STRUCTURE
# ----------------------------------

RESPONSE_STRUCTURE = "ResponseCode | EncryptedPassword"

SAMPLE_RESPONSE = "100|UWQ/yLAS7KmiHyExtsKKq3DuHTWuyGOzy0P4s38PSL8nA=="

print("Response Format:", RESPONSE_STRUCTURE)
print("Example Response:", SAMPLE_RESPONSE)
```

```python
# ----------------------------------
# RESPONSE CODES
# ----------------------------------

RESPONSE_CODES = {
    "100": "Login Successful",
    "101": "Invalid Login"
}

RESPONSE_CODES
```

## Purpose of this API

This API authenticates the member system with BSE Star MF.

If authentication is successful:

Response code = **100**

An **Encrypted Password** is generated.

This encrypted password acts as a **Session ID** and must be used for further API requests such as:

- Order Entry
- SIP Orders
- Switch Orders
- Redemption
