# XSIP / ISIP Order Entry API – Python Reference

This notebook provides the **Python reference structure** for the **XSIP / ISIP Order Entry API** used in the **BSE StAR MF Web Services**.

**API Method:** `xsipOrderEntryParam`

## Purpose

This API allows members to **register or cancel an XSIP or ISIP (Extended / Instant Systematic Investment Plan)** for mutual fund investments on behalf of clients.

XSIP and ISIP allow investors to invest a **fixed amount at regular intervals** in mutual funds using a **mandate-based payment system**.

The request is sent from the **member system (for example FundsPi platform)** to the **BSE StAR MF server** using **SOAP Web Services**.


## Process Overview

The process of placing an XSIP / ISIP order using this API follows these steps:

1. The member system first **authenticates using the Authentication API**.
2. The system **generates an encrypted password and passkey** required for secure communication.
3. The **XSIP / ISIP order request is prepared** with required parameters such as:
   - Transaction Code  
   - Unique Reference Number  
   - Scheme Code  
   - Client Code  
   - Installment Amount  
   - Frequency Type  
   - Mandate ID  
4. The request data is **converted into SOAP XML format**.
5. The SOAP request is **sent to the BSE StAR MF server endpoint**.
6. The BSE server **validates the request and processes the order**.
7. If the request is successful, the server **registers the XSIP / ISIP** and generates an **XSIP Registration ID**.
8. The server sends a **SOAP response message** containing the **registration number and status message**.

```python
# API BASIC INFORMATION

API_NAME = "XSIP / ISIP Order Entry API"

API_METHOD = "xsipOrderEntryParam"

API_ENDPOINT = "https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure"

SOAP_ACTION = "http://bsestarmf.in/MFOrderEntry/xsipOrderEntryParam"

print("API Name:", API_NAME)
print("API Method:", API_METHOD)
print("API Endpoint:", API_ENDPOINT)
```

# Sample SOAP Request:- 

```python
SOAP_REQUEST_SAMPLE = """
<soap:Envelope xmlns:soap="http://www.w3.org/2003/05/soap-envelope"
xmlns:bses="http://bsestarmf.in/">

<soap:Header xmlns:wsa="http://www.w3.org/2005/08/addressing">
<wsa:Action>http://bsestarmf.in/MFOrderEntry/xsipOrderEntryParam</wsa:Action>

<wsa:To>
https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure
</wsa:To>

</soap:Header>

<soap:Body>

<bses:xsipOrderEntryParam>

<bses:TransactionCode>NEW</bses:TransactionCode>
<bses:UniqueRefNo>2023070210234000001</bses:UniqueRefNo>
<bses:SchemeCode>QUOFGP-GR</bses:SchemeCode>
<bses:MemberCode>1234</bses:MemberCode>
<bses:ClientCode>clientcode01</bses:ClientCode>

</bses:xsipOrderEntryParam>

</soap:Body>

</soap:Envelope>
"""

print(SOAP_REQUEST_SAMPLE)
```
