# XSIP / ISIP Order Entry API Structure

This notebook documents the structure of the XSIP / ISIP Order Entry API used in BSE Star MF Web Services.

API Method:
**xsipOrderEntryParam**

Purpose:
This API is used to register or cancel an XSIP or ISIP order for mutual fund investments.

The request is sent from the member system to the BSE Star MF server using SOAP Web Services.

# Introduction to XSIP / ISIP

XSIP and ISIP allow investors to invest a fixed amount in mutual funds at regular intervals.

Examples of SIP frequency:

• Monthly  
• Quarterly  
• Weekly
