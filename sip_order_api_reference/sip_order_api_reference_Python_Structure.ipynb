# SIP Order Entry API – Python Reference

This notebook provides the Python reference structure for the **SIP Order Entry API** used in the BSE StAR MF Web Services.

API Method:
**sipOrderEntryParam**

Purpose:
This API allows members to create or cancel a SIP (Systematic Investment Plan) for mutual fund investments on behalf of clients.

The request is sent from the member system to the BSE StAR MF server using SOAP Web Services.

Process Overview:

1. The member system authenticates using the Authentication API.
2. The system generates an encrypted password and passkey.
3. The SIP order request is prepared with required parameters.
4. The request is sent to the BSE StAR MF server.
5. The server validates the request and registers the SIP.
6. A response is returned containing the SIP registration ID and confirmation message.

This notebook contains:

• API basic information  
• Request parameter structure  
• Response structure  
• Sample SOAP request  
• Sample SOAP response

```python
# -------------------------------------
# API BASIC INFORMATION
# -------------------------------------

API_NAME = "SIP Order Entry API"

API_METHOD = "sipOrderEntryParam"

API_ENDPOINT = "https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure"

SOAP_ACTION = "http://bsestarmf.in/MFOrderEntry/sipOrderEntryParam"

print("API Name:", API_NAME)
print("API Method:", API_METHOD)
print("API Endpoint:", API_ENDPOINT)
```

```python
# -------------------------------------
# SAMPLE SOAP REQUEST
# -------------------------------------

SOAP_REQUEST_SAMPLE = """
<soap:Envelope xmlns:soap="http://www.w3.org/2003/05/soap-envelope"
xmlns:bses="http://bsestarmf.in/">

<soap:Header xmlns:wsa="http://www.w3.org/2005/08/addressing">
<wsa:Action>http://bsestarmf.in/MFOrderEntry/sipOrderEntryParam</wsa:Action>

<wsa:To>
https://bsestarmfdemo.bseindia.com/MFOrderEntry/MFOrder.svc/Secure
</wsa:To>

</soap:Header>

<soap:Body>

<bses:sipOrderEntryParam>

<bses:TransactionCode>NEW</bses:TransactionCode>
<bses:UniqueRefNo>20230810000011</bses:UniqueRefNo>
<bses:SchemeCode>B301G</bses:SchemeCode>

</bses:sipOrderEntryParam>

</soap:Body>

</soap:Envelope>
"""

print(SOAP_REQUEST_SAMPLE)
```

```python
# -------------------------------------
# SAMPLE SOAP RESPONSE
# -------------------------------------

SOAP_RESPONSE_SAMPLE = """
<sipOrderEntryParamResponse>

<sipOrderEntryParamResult>

NEW|20230810000011|1234|clientcode01|1234|582596|
SIP HAS BEEN REGISTERED, SIP REG NO IS : 582596|0|0

</sipOrderEntryParamResult>

</sipOrderEntryParamResponse>
"""

print(SOAP_RESPONSE_SAMPLE)
```
