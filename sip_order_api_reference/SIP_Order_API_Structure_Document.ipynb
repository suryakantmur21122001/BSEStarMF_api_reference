# SIP Order Entry API Structure

This notebook documents the structure of the **SIP Order Entry API** used in BSE Star MF Web Services.

API Method:
**sipOrderEntryParam**

Purpose:
This API is used to create or cancel a SIP (Systematic Investment Plan) order for a mutual fund investment.

The SIP order is submitted from the member system to the BSE Star MF server using SOAP Web Service.

# Introduction to SIP

A Systematic Investment Plan (SIP) allows investors to invest a fixed amount in mutual funds at regular intervals.

Examples of SIP frequency:

• Monthly SIP  
• Quarterly SIP  
• Weekly SIP
